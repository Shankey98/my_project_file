```python
def kakao_token_first_issue(CODE, REST_API_KEY, REDIRECT_URI, CLIENT_SECRET, save_path):
    url = "https://kauth.kakao.com/oauth/token"
    data = {
        "grant_type": "authorization_code",
        "client_id": REST_API_KEY,
        "redirect_uri": REDIRECT_URI,
        "code": CODE,
        "client_secret": CLIENT_SECRET
    }
    response = requests.post(url, data=data)
    tokens = response.json()
    print(tokens)
    with open(save_path, "w") as fp:
        json.dump(tokens, fp)
# ê²°ê³¼: {'access_token': '...', 'refresh_token': '...', ...}
```

```python
def send_kakao_me(access_token, hope_date, link):
    url = "https://kapi.kakao.com/v2/api/talk/memo/default/send"
    headers = {
        "Authorization": f"Bearer {access_token}"
    }
    
    # ì¼ë°˜ í…ìŠ¤íŠ¸ì— URL í¬í•¨ - ì¹´ì¹´ì˜¤í†¡ì´ ìë™ìœ¼ë¡œ ë¯¸ë¦¬ë³´ê¸° ìƒì„±
    message_text = f"""í•œì–‘ëŒ€ ë„ì„œê´€ ì˜ˆì•½ ì™„ë£Œ! ğŸ“š

ğŸ“… ì˜ˆì•½ì¼: {hope_date}
âœ… ì‹œì„¤ì˜ˆì•½ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.

ğŸ“‹ ì˜ˆì•½ë‚´ì—­ í™•ì¸í•˜ê¸°:
{link}"""
    
    data = {
        "template_object": json.dumps({
            "object_type": "text",
            "text": message_text,
            "link": {
                "web_url": link,
                "mobile_web_url": link
            }
        })
    }
    
    resp = requests.post(url, headers=headers, data=data)
    print(resp.json())
    
    if resp.json().get('result_code') == 0:
        print('ë©”ì‹œì§€ë¥¼ ì„±ê³µì ìœ¼ë¡œ ë³´ëƒˆìŠµë‹ˆë‹¤.')
    else:
        print(f'ë©”ì‹œì§€ë¥¼ ì„±ê³µì ìœ¼ë¡œ ë³´ë‚´ì§€ ëª»í–ˆìŠµë‹ˆë‹¤. ì˜¤ë¥˜ë©”ì‹œì§€ : {resp.json()}')
    
    return resp.json()
```
